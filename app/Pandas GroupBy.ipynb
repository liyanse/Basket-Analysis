```python
import pandas as pd
```

```python
## Read the Dataset
df =pd.read_csv("Groceries_dataset.csv")
df.head()
```

```python
## Single group column
df.groupby(['Member_number'])['itemDescription'].sum()
```

```python
## Double Group column
df.groupby(['Member_number'])[['Date','itemDescription']].sum()
```

```python
df.groupby(['Member_number', 'Date'])['itemDescription'].sum()
```

```python
df.groupby(['Member_number','Date'], as_index = False)['itemDescription'].sum()
```
